## Araç Kiralama Sistemi

### 1. Temel Sınıf: Araç (Vehicle)

```python
# Temel araç sınıfı: Tüm araçların ortak özelliklerini ve davranışlarını tanımlar.
class Araç:
    # Her araç nesnesi oluşturulduğunda çağrılır.
    # plaka, marka, model, yıl ve günlük_fiyat gibi temel bilgileri alır.
    # müsait özniteliği, aracın kiralanabilir olup olmadığını gösterir (başlangıçta True).
    def __init__(self, plaka, marka, model, yıl, günlük_fiyat):
        self.plaka = plaka
        self.marka = marka
        self.model = model
        self.yıl = yıl
        self.günlük_fiyat = günlük_fiyat
        self.müsait = True

    # Aracın bilgilerini okunabilir bir string formatında döndürür.
    def display(self):
        return f"Plaka: {self.plaka}, Marka: {self.marka}, Model: {self.model}, Yıl: {self.yıl}, Günlük Fiyat: {self.günlük_fiyat} TL, Müsait: {self.müsait}"
```

### 2. Alt Sınıflar: Otomobil (Car) ve Motosiklet (Motorcycle)

```python
# Araç sınıfından türetilmiş Otomobil sınıfı.
# Otomobile özgü ek özellikler (kapı_sayısı, bagaj_hacmi) içerir.
class Otomobil(Araç):
    # Üst sınıfın (Araç) başlatıcısını çağırır ve otomobile özel öznitelikleri ekler.
    def __init__(self, plaka, marka, model, yıl, günlük_fiyat, kapı_sayısı, bagaj_hacmi):
        super().__init__(plaka, marka, model, yıl, günlük_fiyat)
        self.kapı_sayısı = kapı_sayısı
        self.bagaj_hacmi = bagaj_hacmi

    # Otomobilin bilgilerini, temel araç bilgilerine ek olarak kendi özel bilgilerini de içerecek şekilde döndürür.
    def display(self):
        base_info = super().display()
        return f"{base_info}, Kapı Sayısı: {self.kapı_sayısı}, Bagaj Hacmi: {self.bagaj_hacmi} L"

# Araç sınıfından türetilmiş Motosiklet sınıfı.
# Motosiklete özgü ek özellikler (motor_hacmi) içerir.
class Motosiklet(Araç):
    # Üst sınıfın (Araç) başlatıcısını çağırır ve motosiklete özel özniteliği ekler.
    def __init__(self, plaka, marka, model, yıl, günlük_fiyat, motor_hacmi):
        super().__init__(plaka, marka, model, yıl, günlük_fiyat)
        self.motor_hacmi = motor_hacmi

    # Motosikletin bilgilerini, temel araç bilgilerine ek olarak kendi özel bilgilerini de içerecek şekilde döndürür.
    def display(self):
        base_info = super().display()
        return f"{base_info}, Motor Hacmi: {self.motor_hacmi} cc"
```

### 3. Müşteri Sınıfı (Customer Class)

```python
# Müşteri sınıfı: Kiralama yapacak müşterilerin bilgilerini yönetir.
class Müşteri:
    # Her müşteri nesnesi oluşturulduğunda çağrılır.
    # müşteri_id, ad_soyad ve telefon_numarası gibi bilgileri alır.
    # kiralanan_araçlar listesi, müşterinin şu anda kiraladığı araçların plakalarını tutar.
    def __init__(self, müşteri_id, ad_soyad, telefon_numarası):
        self.müşteri_id = müşteri_id
        self.ad_soyad = ad_soyad
        self.telefon_numarası = telefon_numarası
        self.kiralanan_araçlar = [] # Müşterinin kiraladığı araçların plakaları tutulur.

    # Müşterinin bilgilerini okunabilir bir string formatında döndürür.
    def display(self):
        return f"Müşteri ID: {self.müşteri_id}, Ad Soyad: {self.ad_soyad}, Telefon: {self.telefon_numarası}, Kiralanan Araç Sayısı: {len(self.kiralanan_araçlar)}"
```

### 4. Kiralama Sistemi Sınıfı (RentalSystem Class)

```python
import datetime

class KiralamaSistemi:
    def __init__(self):
        self.araçlar = {}
        self.müşteriler = {}
        self.kiralama_kayıtları = [] # (müşteri_id, plaka, kiralama_tarihi, iade_tarihi, ücret)

    def araç_ekle(self, araç):
        self.araçlar[araç.plaka] = araç
        print(f"{araç.plaka} plakalı araç sisteme eklendi.")

    def müşteri_ekle(self, müşteri):
        self.müşteriler[müşteri.müşteri_id] = müşteri
        print(f"{müşteri.ad_soyad} isimli müşteri sisteme eklendi.")

    def araçları_listele(self):
        if not self.araçlar:
            print("Sistemde kayıtlı araç bulunmamaktadır.")
            return
        print("\nTüm Araçlar:")
        for araç in self.araçlar.values():
            print(araç.display())

    def müsait_araçları_listele(self):
        müsait_olanlar = [araç for araç in self.araçlar.values() if araç.müsait]
        if not müsait_olanlar:
            print("Şu anda müsait araç bulunmamaktadır.")
            return
        print("\nMüsait Araçlar:")
        for araç in müsait_olanlar:
            print(araç.display())

    def araç_kirala(self, müşteri_id, plaka, kiralama_tarihi_str):
        müşteri = self.müşteriler.get(müşteri_id)
        araç = self.araçlar.get(plaka)

        if not müşteri:
            print(f"Hata: {müşteri_id} ID'li müşteri bulunamadı.")
            return False
        if not araç:
            print(f"Hata: {plaka} plakalı araç bulunamadı.")
            return False
        if not araç.müsait:
            print(f"Hata: {plaka} plakalı araç müsait değil.")
            return False

        try:
            kiralama_tarihi = datetime.datetime.strptime(kiralama_tarihi_str, "%Y-%m-%d")
        except ValueError:
            print("Hata: Kiralama tarihi formatı yanlış. YYYY-MM-DD formatını kullanın.")
            return False

        araç.müsait = False
        self.kiralama_kayıtları.append((müşteri_id, plaka, kiralama_tarihi, None, 0))
        müşteri.kiralanan_araçlar.append(plaka)
        print(f"{plaka} plakalı araç, {müşteri.ad_soyad} tarafından {kiralama_tarihi_str} tarihinde kiralandı.")
        return True

    def araç_iade_et(self, müşteri_id, plaka, iade_tarihi_str):
        müşteri = self.müşteriler.get(müşteri_id)
        araç = self.araçlar.get(plaka)

        if not müşteri:
            print(f"Hata: {müşteri_id} ID'li müşteri bulunamadı.")
            return False, 0
        if not araç:
            print(f"Hata: {plaka} plakalı araç bulunamadı.")
            return False, 0
        if araç.müsait:
            print(f"Hata: {plaka} plakalı araç zaten müsait durumda (kiralanmamış olabilir).")
            return False, 0
        if plaka not in müşteri.kiralanan_araçlar:
            print(f"Hata: {müşteri.ad_soyad} bu {plaka} plakalı aracı kiralamamış.")
            return False, 0

        try:
            iade_tarihi = datetime.datetime.strptime(iade_tarihi_str, "%Y-%m-%d")
        except ValueError:
            print("Hata: İade tarihi formatı yanlış. YYYY-MM-DD formatını kullanın.")
            return False, 0

        kiralama_kaydı_index = -1
        for i, kayıt in enumerate(self.kiralama_kayıtları):
            if kayıt[0] == müşteri_id and kayıt[1] == plaka and kayıt[3] is None:
                kiralama_kaydı_index = i
                break

        if kiralama_kaydı_index == -1:
            print(f"Hata: {müşteri.ad_soyad} için {plaka} plakalı aktif kiralama kaydı bulunamadı.")
            return False, 0

        kiralama_tarihi = self.kiralama_kayıtları[kiralama_kaydı_index][2]
        if iade_tarihi < kiralama_tarihi:
            print("Hata: İade tarihi kiralama tarihinden önce olamaz.")
            return False, 0

        kiralama_süresi = (iade_tarihi - kiralama_tarihi).days
        if kiralama_süresi == 0: # Same day rental counts as 1 day
            kiralama_süresi = 1

        ücret = self.kiralama_ücretini_hesapla(plaka, kiralama_süresi)

        # Update the rental record
        self.kiralama_kayıtları[kiralama_kaydı_index] = (müşteri_id, plaka, kiralama_tarihi, iade_tarihi, ücret)
        araç.müsait = True
        müşteri.kiralanan_araçlar.remove(plaka)

        print(f"{plaka} plakalı araç {müşteri.ad_soyad} tarafından iade edildi. Toplam Ücret: {ücret} TL")
        return True, ücret

    def kiralama_ücretini_hesapla(self, plaka, kiralama_süresi):
        araç = self.araçlar.get(plaka)
        if not araç:
            print(f"Hata: {plaka} plakalı araç bulunamadı.")
            return 0

        # Basit ücret hesaplama: günlük_fiyat * kiralama_süresi
        # Farklı araç tipleri için farklı hesaplama politikaları burada genişletilebilir (Polymorphism)
        # Örneğin: if isinstance(araç, Otomobil): return ...
        # if isinstance(araç, Motosiklet): return ...
        return araç.günlük_fiyat * kiralama_süresi

    def müşteri_kiralama_geçmişi(self, müşteri_id):
        müşteri = self.müşteriler.get(müşteri_id)
        if not müşteri:
            print(f"Hata: {müşteri_id} ID'li müşteri bulunamadı.")
            return

        print(f"\n{müşteri.ad_soyad} adlı müşterinin kiralama geçmişi:")
        geçmiş_bulundu = False
        for kayıt in self.kiralama_kayıtları:
            if kayıt[0] == müşteri_id:
                plaka, kiralama_tarihi, iade_tarihi, ücret = kayıt[1], kayıt[2], kayıt[3], kayıt[4]
                iade_bilgisi = iade_tarihi.strftime("%Y-%m-%d") if iade_tarihi else "Henüz iade edilmedi"
                ücret_bilgisi = f"Ücret: {ücret} TL" if iade_tarihi else ""
                print(f" - Araç Plaka: {plaka}, Kiralama Tarihi: {kiralama_tarihi.strftime("%Y-%m-%d")}, İade Tarihi: {iade_bilgisi}, {ücret_bilgisi}")
                geçmiş_bulundu = True

        if not geçmiş_bulundu:
            print("   Kiralama geçmişi bulunamadı.")
```

### 5. Verileri Kaydetme ve Yükleme (Saving and Loading Data)

```python
import datetime
import json

# KiralamaSistemi sınıfı: Tüm araç kiralama operasyonlarını yönetir.
# Araçları, müşterileri ve kiralama kayıtlarını saklar ve işler.
class KiralamaSistemi:
    # Sistem başlatıldığında çağrılır.
    # Boş araçlar, müşteriler ve kiralama_kayıtları sözlüklerini/listelerini başlatır.
    def __init__(self):
        self.araçlar = {} # Plakaya göre araç nesnelerini saklar.
        self.müşteriler = {} # Müşteri ID'sine göre müşteri nesnelerini saklar.
        # Kiralama kayıtları listesi: (müşteri_id, plaka, kiralama_tarihi, iade_tarihi, ücret) şeklinde tuple'lar tutar.
        self.kiralama_kayıtları = []

    # Yeni bir aracı sisteme ekler.
    def araç_ekle(self, araç):
        self.araçlar[araç.plaka] = araç
        print(f"{araç.plaka} plakalı araç sisteme eklendi.")

    # Yeni bir müşteriyi sisteme ekler.
    def müşteri_ekle(self, müşteri):
        self.müşteriler[müşteri.müşteri_id] = müşteri
        print(f"{müşteri.ad_soyad} isimli müşteri sisteme eklendi.")

    # Sistemdeki tüm araçları listeler.
    def araçları_listele(self):
        if not self.araçlar:
            print("Sistemde kayıtlı araç bulunmamaktadır.")
            return
        print("\nTüm Araçlar:")
        for araç in self.araçlar.values():
            print(araç.display())

    # Sistemdeki müsait (kiralanabilir) araçları listeler.
    def müsait_araçları_listele(self):
        müsait_olanlar = [araç for araç in self.araçlar.values() if araç.müsait]
        if not müsait_olanlar:
            print("Şu anda müsait araç bulunmamaktadır.")
            return
        print("\nMüsait Araçlar:")
        for araç in müsait_olanlar:
            print(araç.display())

    # Bir aracı kiralar.
    # Müşteri ve araç varlığını, aracın müsaitliğini kontrol eder.
    # Kiralama tarihini kaydeder ve aracın müsaitlik durumunu günceller.
    def araç_kirala(self, müşteri_id, plaka, kiralama_tarihi_str):
        müşteri = self.müşteriler.get(müşteri_id)
        araç = self.araçlar.get(plaka)

        if not müşteri:
            print(f"Hata: {müşteri_id} ID'li müşteri bulunamadı.")
            return False
        if not araç:
            print(f"Hata: {plaka} plakalı araç bulunamadı.")
            return False
        if not araç.müsait:
            print(f"Hata: {plaka} plakalı araç müsait değil.")
            return False

        try:
            kiralama_tarihi = datetime.datetime.strptime(kiralama_tarihi_str, "%Y-%m-%d")
        except ValueError:
            print("Hata: Kiralama tarihi formatı yanlış. YYYY-MM-DD formatını kullanın.")
            return False

        araç.müsait = False # Aracı müsait değil olarak işaretle
        # Yeni kiralama kaydını listeye ekle (iade tarihi ve ücret başlangıçta boş/sıfır)
        self.kiralama_kayıtları.append((müşteri_id, plaka, kiralama_tarihi, None, 0))
        müşteri.kiralanan_araçlar.append(plaka) # Müşterinin kiralanan araçlar listesine ekle
        print(f"{plaka} plakalı araç, {müşteri.ad_soyad} tarafından {kiralama_tarihi_str} tarihinde kiralandı.")
        return True

    # Kiralanan bir aracı iade alır.
    # Müşteri ve araç varlığını, aracın kiralı olup olmadığını kontrol eder.
    # Kiralama süresini hesaplar, ücreti belirler ve aracın müsaitlik durumunu günceller.
    def araç_iade_et(self, müşteri_id, plaka, iade_tarihi_str):
        müşteri = self.müşteriler.get(müşteri_id)
        araç = self.araçlar.get(plaka)

        if not müşteri:
            print(f"Hata: {müşteri_id} ID'li müşteri bulunamadı.")
            return False, 0
        if not araç:
            print(f"Hata: {plaka} plakalı araç bulunamadı.")
            return False, 0
        if araç.müsait:
            print(f"Hata: {plaka} plakalı araç zaten müsait durumda (kiralanmamış olabilir).")
            return False, 0
        if plaka not in müşteri.kiralanan_araçlar:
            print(f"Hata: {müşteri.ad_soyad} bu {plaka} plakalı aracı kiralamamış.")
            return False, 0

        try:
            iade_tarihi = datetime.datetime.strptime(iade_tarihi_str, "%Y-%m-%d")
        except ValueError:
            print("Hata: İade tarihi formatı yanlış. YYYY-MM-DD formatını kullanın.")
            return False, 0

        kiralama_kaydı_index = -1
        # Aktif kiralama kaydını bul
        for i, kayıt in enumerate(self.kiralama_kayıtları):
            if kayıt[0] == müşteri_id and kayıt[1] == plaka and kayıt[3] is None: # Henüz iade edilmemiş kayıt
                kiralama_kaydı_index = i
                break

        if kiralama_kaydı_index == -1:
            print(f"Hata: {müşteri.ad_soyad} için {plaka} plakalı aktif kiralama kaydı bulunamadı.")
            return False, 0

        kiralama_tarihi = self.kiralama_kayıtları[kiralama_kaydı_index][2]
        if iade_tarihi < kiralama_tarihi:
            print("Hata: İade tarihi kiralama tarihinden önce olamaz.")
            return False, 0

        # Kiralama süresini gün cinsinden hesapla.
        kiralama_süresi = (iade_tarihi - kiralama_tarihi).days
        if kiralama_süresi == 0: # Aynı gün kiralamalar 1 gün sayılır
            kiralama_süresi = 1

        ücret = self.kiralama_ücretini_hesapla(plaka, kiralama_süresi)

        # Kiralama kaydını iade bilgileri ve ücret ile güncelle.
        self.kiralama_kayıtları[kiralama_kaydı_index] = (müşteri_id, plaka, kiralama_tarihi, iade_tarihi, ücret)
        araç.müsait = True # Aracı tekrar müsait olarak işaretle
        müşteri.kiralanan_araçlar.remove(plaka) # Müşterinin kiralanan araçlar listesinden çıkar

        print(f"{plaka} plakalı araç {müşteri.ad_soyad} tarafından iade edildi. Toplam Ücret: {ücret} TL")
        return True, ücret

    # Bir aracın kiralama ücretini hesaplar.
    # Araç tipine (Otomobil/Motosiklet) göre farklı ücretlendirme politikaları uygular (Polymorphism).
    def kiralama_ücretini_hesapla(self, plaka, kiralama_süresi):
        araç = self.araçlar.get(plaka)
        if not araç:
            print(f"Hata: {plaka} plakalı araç bulunamadı.")
            return 0

        # Farklı araç tipleri için farklı hesaplama politikaları (Polymorphism örneği)
        if isinstance(araç, Otomobil):
            return araç.günlük_fiyat * kiralama_süresi * 1.1 # Otomobillere %10 ekstra ücret
        elif isinstance(araç, Motosiklet):
            return araç.günlük_fiyat * kiralama_süresi * 0.9 # Motosikletlere %10 indirim
        else: # Diğer araç türleri için varsayılan ücret
            return araç.günlük_fiyat * kiralama_süresi

    # Belirli bir müşterinin tüm kiralama geçmişini görüntüler.
    def müşteri_kiralama_geçmişi(self, müşteri_id):
        müşteri = self.müşteriler.get(müşteri_id)
        if not müşteri:
            print(f"Hata: {müşteri_id} ID'li müşteri bulunamadı.")
            return

        print(f"\n{müşteri.ad_soyad} adlı müşterinin kiralama geçmişi:")
        geçmiş_bulundu = False
        for kayıt in self.kiralama_kayıtları:
            if kayıt[0] == müşteri_id:
                plaka, kiralama_tarihi, iade_tarihi, ücret = kayıt[1], kayıt[2], kayıt[3], kayıt[4]
                iade_bilgisi = iade_tarihi.strftime("%Y-%m-%d") if iade_tarihi else "Henüz iade edilmedi"
                ücret_bilgisi = f", Ücret: {ücret} TL" if iade_tarihi else ""
                print(f" - Araç Plaka: {plaka}, Kiralama Tarihi: {kiralama_tarihi.strftime("%Y-%m-%d")}, İade Tarihi: {iade_bilgisi}{ücret_bilgisi}")
                geçmiş_bulundu = True

        if not geçmiş_bulundu:
            print("   Kiralama geçmişi bulunamadı.")

    # Sistemdeki tüm verileri (araçlar, müşteriler, kiralama kayıtları) bir JSON dosyasına kaydeder.
    # Nesneleri serileştirmek için vars() ve datetime objelerini stringe dönüştürme kullanılır.
    def verileri_kaydet(self, dosya_adı="kiralama_verileri.json"):
        data = {
            "araçlar": {plaka: vars(araç) for plaka, araç in self.araçlar.items()},
            "müşteriler": {id: vars(müşteri) for id, müşteri in self.müşteriler.items()},
            "kiralama_kayıtları": [
                (k[0], k[1], k[2].strftime("%Y-%m-%d"), (k[3].strftime("%Y-%m-%d") if k[3] else None), k[4])
                for k in self.kiralama_kayıtları
            ]
        }
        # Serileştirme sırasında sınıf tiplerini ve datetime nesnelerini yönetmek için dönüştürme
        for plaka, araç_data in data["araçlar"].items():
            # Her araç nesnesinin tipini (Otomobil, Motosiklet, Araç) kaydeder.
            if isinstance(self.araçlar[plaka], Otomobil):
                araç_data['tip'] = 'Otomobil'
            elif isinstance(self.araçlar[plaka], Motosiklet):
                araç_data['tip'] = 'Motosiklet'
            else:
                araç_data['tip'] = 'Araç'

        # Müşteri kiralanan araçlar listesindeki araç objeleri yerine sadece plaka stringlerini kaydet
        # (JSON'da nesne referansları yerine basit stringler saklamak için)
        for müşteri_id, müşteri_data in data["müşteriler"].items():
            müşteri_data['kiralanan_araçlar'] = [plaka for plaka in müşteri_data['kiralanan_araçlar']]

        with open(dosya_adı, 'w', encoding='utf-8') as f: # Türkçe karakter desteği için encoding eklendi
            json.dump(data, f, indent=4, ensure_ascii=False)
        print(f"Veriler '{dosya_adı}' dosyasına kaydedildi.")

    # Bir JSON dosyasından kaydedilmiş verileri (araçlar, müşteriler, kiralama kayıtları) sisteme yükler.
    # Serileştirilmiş verileri orijinal nesne yapılarına geri dönüştürür (deserialization).
    def verileri_yükle(self, dosya_adı="kiralama_verileri.json"):
        try:
            with open(dosya_adı, 'r', encoding='utf-8') as f: # Türkçe karakter desteği için encoding eklendi
                data = json.load(f)

            self.araçlar = {} # Mevcut araçları temizle
            for plaka, araç_data in data["araçlar"].items():
                tip = araç_data.pop('tip') # Kaydedilen araç tipini al
                # Tipe göre doğru araç sınıfını kullanarak nesneyi yeniden oluştur
                if tip == 'Otomobil':
                    self.araçlar[plaka] = Otomobil(araç_data['plaka'], araç_data['marka'], araç_data['model'],
                                                  araç_data['yıl'], araç_data['günlük_fiyat'],
                                                  araç_data['kapı_sayısı'], araç_data['bagaj_hacmi'])
                elif tip == 'Motosiklet':
                    self.araçlar[plaka] = Motosiklet(araç_data['plaka'], araç_data['marka'], araç_data['model'],
                                                    araç_data['yıl'], araç_data['günlük_fiyat'],
                                                    araç_data['motor_hacmi'])
                else: # Generic Araç
                    self.araçlar[plaka] = Araç(araç_data['plaka'], araç_data['marka'], araç_data['model'],
                                              araç_data['yıl'], araç_data['günlük_fiyat'])
                self.araçlar[plaka].müsait = araç_data['müsait'] # Müsaitlik durumunu da yükle

            self.müşteriler = {} # Mevcut müşterileri temizle
            for id, müşteri_data in data["müşteriler"].items():
                # Müşteri nesnesini yeniden oluştur
                müşteri = Müşteri(müşteri_data['müşteri_id'], müşteri_data['ad_soyad'], müşteri_data['telefon_numarası'])
                müşteri.kiralanan_araçlar = müşteri_data['kiralanan_araçlar'] # Kiralanan araçları plaka olarak yükle
                self.müşteriler[id] = müşteri

            self.kiralama_kayıtları = [] # Mevcut kiralama kayıtlarını temizle
            for k in data["kiralama_kayıtları"]:
                # Tarih stringlerini datetime objelerine dönüştürerek kiralama kayıtlarını yeniden oluştur
                kiralama_tarihi = datetime.datetime.strptime(k[2], "%Y-%m-%d")
                iade_tarihi = datetime.datetime.strptime(k[3], "%Y-%m-%d") if k[3] else None
                self.kiralama_kayıtları.append((k[0], k[1], kiralama_tarihi, iade_tarihi, k[4]))

            print(f"Veriler '{dosya_adı}' dosyasından yüklendi.")
        except FileNotFoundError:
            print(f"Hata: '{dosya_adı}' dosyası bulunamadı. Yeni bir sistem başlatılıyor.")
        except Exception as e:
            print(f"Veri yüklenirken bir hata oluştu: {e}")
```

```python
# Ana programın başlangıç noktası.
# Kiralama sistemini başlatır ve kullanıcı etkileşimini yöneten menüyü gösterir.
def main():
    sistem = KiralamaSistemi()

    # Başlangıç test verileri eklenir.
    sistem.araç_ekle(Otomobil("34 ABC 123", "Ford", "Focus", 2020, 300, 4, 350))
    sistem.araç_ekle(Otomobil("06 DEF 456", "Opel", "Corsa", 2018, 250, 2, 280))
    sistem.araç_ekle(Motosiklet("45 MOTO 789", "Yamaha", "MT-07", 2022, 400, 689))
    sistem.araç_ekle(Motosiklet("10 BIKE 001", "Honda", "CBR500R", 2021, 350, 471))

    sistem.müşteri_ekle(Müşteri("M001", "Ayşe Yılmaz", "5321112233"))
    sistem.müşteri_ekle(Müşteri("M002", "Can Demir", "5434445566"))

    # Ana menü döngüsü: Kullanıcı çıkış yapana kadar menüyü gösterir ve seçimleri işler.
    while True:
        print("\nARAÇ KİRALAMA SİSTEMİ")
        print("1- Araç ekle")
        print("2- Müşteri ekle")
        print("3- Araçları listele")
        print("4- Müsait araçları listele")
        print("5- Araç kirala")
        print("6- Araç iade et")
        print("7- Kiralama ücretini hesapla (sadece bilgi amaçlı)")
        print("8- Müşteri kiralama geçmişi")
        print("9- Verileri kaydet")
        print("10- Verileri yükle")
        print("0- Çıkış")

        seçim = input("Seçiminizi yapın: ")

        # Kullanıcının seçimine göre ilgili KiralamaSistemi metodunu çağırır veya programı sonlandırır.
        if seçim == '1': # Araç ekleme işlemi
            tip = input("Eklenecek araç tipi (Otomobil/Motosiklet): ").lower()
            plaka = input("Plaka: ")
            marka = input("Marka: ")
            model = input("Model: ")
            yıl = int(input("Yıl: "))
            günlük_fiyat = float(input("Günlük Fiyat: "))
            if tip == 'otomobil':
                kapı_sayısı = int(input("Kapı Sayısı: "))
                bagaj_hacmi = float(input("Bagaj Hacmi (L): "))
                sistem.araç_ekle(Otomobil(plaka, marka, model, yıl, günlük_fiyat, kapı_sayısı, bagaj_hacmi))
            elif tip == 'motosiklet':
                motor_hacmi = int(input("Motor Hacmi (cc): "))
                sistem.araç_ekle(Motosiklet(plaka, marka, model, yıl, günlük_fiyat, motor_hacmi))
            else:
                print("Geçersiz araç tipi.")

        elif seçim == '2': # Müşteri ekleme işlemi
            müşteri_id = input("Müşteri ID: ")
            ad_soyad = input("Ad Soyad: ")
            telefon_numarası = input("Telefon Numarası: ")
            sistem.müşteri_ekle(Müşteri(müşteri_id, ad_soyad, telefon_numarası))

        elif seçim == '3': # Tüm araçları listeleme
            sistem.araçları_listele()

        elif seçim == '4': # Müsait araçları listeleme
            sistem.müsait_araçları_listele()

        elif seçim == '5': # Araç kiralama işlemi
            müşteri_id = input("Kiralamak isteyen Müşteri ID: ")
            plaka = input("Kiralanacak Araç Plakası: ")
            kiralama_tarihi_str = input("Kiralama Tarihi (YYYY-MM-DD): ")
            sistem.araç_kirala(müşteri_id, plaka, kiralama_tarihi_str)

        elif seçim == '6': # Araç iade etme işlemi
            müşteri_id = input("Aracı iade eden Müşteri ID: ")
            plaka = input("İade edilecek Araç Plakası: ")
            iade_tarihi_str = input("İade Tarihi (YYYY-MM-DD): ")
            sistem.araç_iade_et(müşteri_id, plaka, iade_tarihi_str)

        elif seçim == '7': # Kiralama ücreti hesaplama (simülasyon amaçlı)
            plaka = input("Ücreti hesaplanacak aracın plakası: ")
            gün_sayısı = int(input("Kaç günlük kiralama (simülasyon için): "))
            ücret = sistem.kiralama_ücretini_hesapla(plaka, gün_sayısı)
            print(f"{plaka} plakalı araç için {gün_sayısı} günlük kiralama ücreti: {ücret} TL")

        elif seçim == '8': # Müşteri kiralama geçmişini görüntüleme
            müşteri_id = input("Geçmişi görüntülenecek Müşteri ID: ")
            sistem.müşteri_kiralama_geçmişi(müşteri_id)

        elif seçim == '9': # Verileri dosyaya kaydetme
            sistem.verileri_kaydet()

        elif seçim == '10': # Verileri dosyadan yükleme
            sistem.verileri_yükle()

        elif seçim == '0': # Programdan çıkış
            print("Sistemden çıkılıyor...")
            break

        else: # Geçersiz menü seçimi
            print("Geçersiz seçim. Lütfen tekrar deneyin.")

# main() fonksiyonunu yalnızca dosya doğrudan çalıştırıldığında çağırır.
# Bu, kodun bir modül olarak içe aktarıldığında otomatik olarak çalışmasını engeller.
if __name__ == "__main__":
    main()
```
